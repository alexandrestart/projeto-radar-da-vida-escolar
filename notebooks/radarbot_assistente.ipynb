# Aula 06 - RadarBot: Assistente do Radar da Vida Escolar

Este notebook cria uma primeira versão de um assistente de IA baseado nos dados do projeto Radar da Vida Escolar.

O objetivo do assistente é responder perguntas sobre os principais padrões encontrados na pesquisa e sugerir estratégias simples de organização escolar.

```python
from google import genai
from google.colab import userdata

api_key = userdata.get("GOOGLE_API_KEY")

client = genai.Client(api_key=api_key)

modelo = "gemini-2.5-flash"

print("Gemini configurado com sucesso!")
```

# Aula 07 - Base contexto e prompt de sistema

```python
base_contexto = """
  Projeto: Radar da Vida Escolar.

  O projeto investigou dificuldades dos estudantes na organização da rotina escolar.

  A pesquisa analisou dados como:
  - série atual;
  - turno de estudo;
  - canais usados para acompanhar avisos;
  - horas semanais de estudo;
  - frequência de uso de agenda;
  - dificuldade com tarefas;
  - relatos abertos sobre a maior dificuldade da rotina.

  Principais temas encontrados:
  - procrastinação;
  - distrações com celular e redes sociais;
  - falta de cronograma;
  - acúmulo de tarefas;
  - dificuldade de foco;
  - dificuldade para conciliar atividades extras e estudos.

  O assistente deve usar essas informações para explicar padrões e sugerir estratégias simples de organização escolar.

  Limites:
  - não inventar dados;
  - não identificar alunos;
  - não fazer diagnósticos;
  - não afirmar que a pesquisa prova algo absoluto.
"""
```

```python
prompt_sistema = """
  Você é o RadarBot, um assistente escolar criado para ajudar estudantes a entenderem dados sobre organização da rotina escolar.

  Use a base de contexto fornecida para responder às perguntas.

  Responda com linguagem simples, direta e acolhedora.

  Sempre que possível, relacione sua resposta aos dados e temas do projeto.

  Se não houver informação suficiente, diga que os dados disponíveis não permitem responder com certeza.
"""
```

```python
def perguntar_radarbot(pergunta):
  prompt_completo = f"""
    {prompt_sistema}

    Base de contexto:
    {base_contexto}

    Pergunta do usuário:
    {pergunta}
  """

  resposta = client.models.generate_content(
      model=modelo,
      contents=prompt_completo
  )

  return resposta.text
```

```python
pergunta_usuario = "Quais foram os principais problemas encontrados na rotina escolar?"

resposta = perguntar_radarbot(pergunta_usuario)

print(resposta)
```

# Aula 08 - Teste de segurança e finalização do protótipo

```python
guardrails = """
Regras de segurança do assistente:

1. Responda apenas sobre assuntos relacionados ao projeto e à base de contexto.
2. Não invente números, porcentagens, médias ou dados que não foram fornecidos.
3. Não identifique alunos individualmente.
4. Não faça diagnósticos médicos, emocionais ou psicológicos.
5. Não use linguagem ofensiva, julgadora ou discriminatória.
6. Não diga que a pesquisa prova algo absoluto.
7. Quando não houver informação suficiente, responda:
"Os dados disponíveis não permitem responder isso com certeza."
8. Use linguagem simples, respeitosa e acolhedora.
"""
```

```python
# ATUALIZAR A FUNÇÃO
def perguntar_radarbot(pergunta):
  prompt_completo = f"""
    {prompt_sistema}

    Base de contexto:
    {base_contexto}

    Guardrails:
    {guardrails}

    Pergunta do usuário:
    {pergunta}
  """

  resposta = client.models.generate_content(
      model=modelo,
      contents=prompt_completo
  )

  return resposta.text
```
