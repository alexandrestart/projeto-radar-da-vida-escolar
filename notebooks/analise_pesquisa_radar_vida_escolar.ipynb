# Aula 3 - Limpando os dados

```python
# criar uma cópia da base

df_limpo = df.copy()
df_limpo.head()
```

```python
# verificar valores vazios
df_limpo.isnull().sum()
```

```python
# verificar linhas duplicas
df_limpo.duplicated().sum()
```

```python
df_limpo["canal_avisos"] = df_limpo["canal_avisos"].replace({
    "Whatsapp": "whatsapp",
    "Site da escola ou portal": "site_portal",
    "Aplicativo da escola ou portal": "app_portal"
})
df_limpo["canal_avisos"].head()
```

```python
# conferir as categorias da coluna de avisos
df_limpo["canal_avisos"].value_counts()
```

```python
# converter colunas de tempo para horas (numérico)

# df_limpo["tempo_diario_escola"] = df_limpo["tempo_diario_escola"].dt.total_seconds()
df_limpo["tempo_diario_escola"] = df_limpo["tempo_diario_escola"] / 3600
df_limpo["tempo_diario_escola"]
```

```python
df_limpo["horas_semana_estudo"] = df_limpo["horas_semana_estudo"].dt.total_seconds() / 3600
df_limpo["horas_semana_estudo"]
```

```python
# mostrar os dados atualizados
df_limpo[["tempo_diario_escola", "horas_semana_estudo"]].head()
```

```python
df_limpo.info()
```

# Aula 4 - Selecionando e filtrando dados

```python
df_limpo.head()
```

```python
colunas_analise = [
    "ano_serie_atual",
    "turno_escola",
    "canal_avisos",
    "horas_semana_estudo",
    "dificuldade_tarefas",
    "frequencia_agenda"
]
```

```python
df_recorte = df_limpo[colunas_analise].copy()
df_recorte.head()
```

```python
# recorte por série
# verificação --> quais dados da minha coluna ano_serie_atual == 1ª Série do Ensino Médio
df_1serie = df_recorte[df_recorte["ano_serie_atual"] == "1ª Série do Ensino Médio"]
df_1serie
```

```python
# recorte por pouco tempo de estudo
df_pouco_tempo = df_recorte[df_recorte["horas_semana_estudo"] < 5]
df_pouco_tempo
```

```python
# recorte combinado
# alunos que estudam menos 5 horas por semana e tem dificuldade alta

df_grupo = df_recorte[
    (df_recorte["horas_semana_estudo"] < 5) &
    (df_recorte["dificuldade_tarefas"] >= 4)
].copy()
len(df_grupo)
```

# Aula 5 - Agrupando e resumindo dados com Pandas

```python
df_recorte.head()
```

```python
# media de horas de estudo
df_recorte["horas_semana_estudo"].mean()
```

```python
# media de dificuldade
df_recorte["dificuldade_tarefas"].mean()
```

```python
# comparar por serie as horas de estudo por semana
df_recorte.groupby("ano_serie_atual")["horas_semana_estudo"].mean()
```

```python
# comparar a dificuldade por serie
df_recorte.groupby("ano_serie_atual")["dificuldade_tarefas"].mean()
```

```python
# a qtd de horas de estudo por semana pelo turno da escola
df_recorte.groupby("turno_escola")["horas_semana_estudo"].mean()
```

```python
df_recorte["ano_serie_atual"].value_counts()
```

```python
# agrupar informações

df_recorte.groupby("ano_serie_atual").agg({
    "horas_semana_estudo": "mean",
    "dificuldade_tarefas": "mean"
})
```

# Aula 6 - Criando gráficos simples

```python
# importar uma bibilioteca de gráficos
import matplotlib.pyplot as plt
```

```python
df_recorte.head()
```

```python
# gráfico de barras: canais de avisos
df_recorte["canal_avisos"].value_counts().plot(kind="bar")

plt.title("Canais de avisos mais utilizados")
plt.xlabel("Canal")
plt.ylabel("Quantidade de alunos")
plt.show()
```

```python
# gráfico de barra: hora de estudo por série
df_recorte.groupby("ano_serie_atual")["horas_semana_estudo"].mean().plot(kind="bar")

plt.title("Média de horas de estudo por série")
plt.xlabel("Ano ou Série")
plt.ylabel("Qtd. de horas")
plt.show()
```

```python
# gráfico de barra: dificuldade por série
df_recorte.groupby("ano_serie_atual")["dificuldade_tarefas"].mean().plot(kind="bar")

plt.title("Dificuldade por série")
plt.xlabel("Ano ou Série")
plt.ylabel("Nível de dificuldade")
plt.show()
```

```python
# grafico de histograma horas de estudo
df_recorte["horas_semana_estudo"].plot(kind="hist")

plt.title("Distribuição das horas de estudo")
plt.xlabel("Horas de estudo por semana")
plt.ylabel("Quantidade de alunos")
plt.show()
```

```python
# gráfico de barra: horas de estudo por canal de aviso
df_recorte.groupby("canal_avisos")["horas_semana_estudo"].mean().plot(kind="bar")

plt.title("Horas de estudo por canal de aviso")
plt.xlabel("Canal")
plt.ylabel("Qtd. de horas de estudo")
plt.show()
```

## Como transformar dados em insight

Um insight segue uma lógica simples:

1. mostrar o dado
2. interpretar o que ele indica

Exemplo:

Dado:
"A média de horas de estudo é baixa"

Insight:
"A baixa quantidade de horas de estudo pode estar relacionada ao aumento da dificuldade nas tarefas"

## Exemplo real com os dados do projeto

Abaixo estão exemplos de insights construídos a partir da base do Radar da Vida Escolar.

### Insight 1 — Tempo de estudo

"A análise mostrou que parte dos alunos apresenta poucas horas de estudo por semana. Esse comportamento pode estar diretamente relacionado ao aumento da dificuldade relatada nas tarefas."

### Insight 2 — Comunicação escolar

"A maioria dos alunos utiliza um único canal principal de comunicação, como o WhatsApp. Isso indica uma forte dependência desse canal e pode gerar problemas caso as informações não estejam centralizadas."

### Insight 3 — Diferença entre séries

"Os dados indicam que algumas séries apresentam maior nível médio de dificuldade. Isso pode estar relacionado à carga de conteúdo ou à adaptação dos alunos a novas exigências acadêmicas."

## Escrever a síntese do projeto

Agora vamos juntar os insights em um pequeno texto.

Exemplo:

"A análise do Radar da Vida Escolar mostrou que a rotina dos alunos é impactada por fatores como baixo tempo de estudo, concentração da comunicação em poucos canais e aumento da dificuldade em determinados grupos. Esses elementos ajudam a explicar os principais desafios enfrentados pelos alunos na organização da vida escolar."
